```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import nltk
from nltk.corpus import stopwords
import math
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
nltk.download('stopwords')
```

```python
text = 'Ferrari S.p.A. is an Italian luxury sports car manufacturer based in Maranello, Italy. Founded in 1939 by Enzo Ferrari, the company adopted its current name in 1945 and began producing its line of cars in 1947. Ferrari became a public company in 1960, and from 1963 to 2014 it was a subsidiary of Fiat S.p.A.'
```

Tokenization

```python
words = text.split()
```

Stop Word Removal

```python
stop_words = set(stopwords.words('english'))
wos = [w for w in words if not w.lower() in stop_words]
wos
```

```python
data = pd.DataFrame({'':wos})
```

Label Encoding

```python
label_encoder = LabelEncoder()
encoded = label_encoder.fit_transform(wos)
encoded
```

One Hot Encoding

```python
ohe = pd.get_dummies( data, columns = [''])
ohe
```

Distinct words

```python
distinct = set(wos)
distinct
```

Bag of Words


```python
sent = 'Ferrari S.p.A. is an Italian luxury sports car manufacturer based in Maranello, Italy'
```

```python
splitted = sent.split()
```

```python
vectorizer = CountVectorizer()
bow_matrix = vectorizer.fit_transform(splitted)
bow_array = bow_matrix.toarray()
feature_names = vectorizer.get_feature_names_out()
print("BoW Matrix:")
print(bow_array)
print("\nFeature Names:")
print(feature_names)
print("Length : ", len(feature_names))
```

TF-IDF

```python
term_freq = {word: 0 for word in wos}

for token in splitted:
    if token in wos:
        term_freq[token] += 1

document_freq = len(splitted)
inverse_document_freq = {word: math.log(document_freq / (term_freq[word] + 1)) for word in wos}

tfidf_scores = {word: term_freq[word] * inverse_document_freq[word] for word in wos}

print("TF-IDF Scores:\n")
for word, score in tfidf_scores.items():
    print(f"'{word}': {score:.4f}")
```

Optional Task

```python
paragraph = ['Ferrari S.p.A. is an Italian luxury sports car manufacturer based in Maranello, Italy.',' Founded in 1939 by Enzo Ferrari, the company adopted its current name in 1945 and began producing its line of cars in 1947.',' Ferrari became a public company in 1960, and from 1963 to 2014 it was a subsidiary of Aditya Kothari S.p.A.']
```

```python
labels = [0, 1, 2]
label_encoder = LabelEncoder()
encoded_labels = label_encoder.fit_transform(labels)
```

```python
vectorizer = CountVectorizer()
word_freq_matrix = vectorizer.fit_transform(paragraph)
print(word_freq_matrix)
```

```python
X_train, X_test, y_train, y_test = train_test_split(word_freq_matrix, encoded_labels, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
predicted_labels = label_encoder.inverse_transform(y_pred)
print("Predicted Labels:", predicted_labels)
```
